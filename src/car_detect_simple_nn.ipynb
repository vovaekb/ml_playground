```
import keras                                                                                                  
from keras.models import Sequential                                                                           
from keras.layers import Dense                                                                                
from keras.optimizers import SGD                                                                              
from keras.utils import np_utils                                                                              
from sklearn.cross_validation import train_test_split                                                         
import numpy as np                                                                                            
import cv2
```

```
train_path = 'CarData/TrainImages'                                                                            
                                                                                                              
# Load training and test data                                                                                 
pos = 'pos-'                                                                                                                                   
neg = 'neg-'                                                                                                  
data = []                                                                                                     
labels = []
```

```
# data processing methods
def sample_path(label, index):                                                                                
    return '{}/{}{}.pgm'.format(train_path, label, index)                                                     
                                                                                                              
def load_img(img_path):                                                                                       
    return cv2.imread(img_path, 0)                                                                            
                                                                                                              
def img_to_features(img):                                                                                     
    return img.flatten() # cv2.resize(img, (50, 20)).flatten()
```

```
# collect data
for i in range(100):                                                                                          
    img = load_img(sample_path(pos, i))                                                                       
    data.append(img_to_features(img))                                                                         
    labels.append(1)                                                                                          
    img = load_img(sample_path(neg, i))                                                                       
    data.append(img_to_features(img))                                                                         
    labels.append(0)
    
#print(data[0])
```

```
#
# preprocess data
#
data = np.array(data) / 255.0                                                                                 
labels = np_utils.to_categorical(labels, 2)                                                                   
#print(labels)                                                                                                
                                                                                                              
# Split data into training and test data sets                                                                 
train_data, test_data, train_labels, test_labels = train_test_split(data, labels, test_size=0.25, random_state=42)
                                                                                                              
#print(train_data.shape)                                                                                      
#print(train_labels[:])
```

```
# Initialize simple neural network                                                            
model = Sequential()                                                                                          
model.add(Dense(768, input_dim=4000, init="uniform", activation="relu")) # 1000                               
model.add(Dense(384, activation="relu"))                                                                      
model.add(Dense(384, activation="relu"))                                                                      
model.add(Dense(2, activation="softmax"))
```

```
# train network
model.compile(optimizer=SGD(lr=0.01), loss="categorical_crossentropy", metrics=["accuracy"]) # "binary_crossentropy"
model.fit(train_data, train_labels, nb_epoch=15, batch_size=64, verbose=1)                                    
                                                                                                             
```

```
# evaluate model on test data
accuracy = model.evaluate(x=test_data, y=test_labels, batch_size=64)                                          
print("Accuracy: {:.3f}".format(accuracy[1]))
```
